# 5.2 Attribute join and spatial join


<br></br>
<font size="3">There are two ways to combine datasets in `geopandas` â€“ attribute joins and spatial joins.
<br></br>
<br></br>
In an attribute join, a `GeoSeries` or `GeoDataFrame` is combined with a regular pandas Series or DataFrame based on a <b>common variable</b>. This is analogous to normal merging or joining in pandas.
<br></br>
<br></br>
In a Spatial Join, observations from to GeoSeries or GeoDataFrames are combined based on their spatial relationship to one another.</font>



```python
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

<br></br>
<font size="3"> Let's play around with those methods using the US lower 48 states example.</font>

```python
states = gpd.read_file("/Users/Ziqi/Desktop/ASU-GIS-322-Fall-19/data/US_lower_48/US_lower_48.shp")
```

```python
states.plot()
```

<br></br>
<font size="3"> First of all, let's calculate the centroid of each state.</font>

```python
states.centroid
```

```python
states.centroid.plot()
```

<br></br>
<font size="3">Overlay state centroids with the state polygons.</font>

```python
fig, ax = plt.subplots(1, 1,figsize=(10,8))
states.plot(ax=ax)
states.centroid.plot(ax=ax,color="red")
```

<br></br>
<font size="3">Then, let's create a buffer to each of the centroid. Before we do that, one important step is to check the projection. If you remember, this one is not projected.</font> 

```python
states.crs
```

```python
states_projected = states.to_crs(epsg='3857')
```

<br></br>
<font size="3">Project it to mercator, so the the buffering distance will be in the unit of meters rather than arc-degree. </font> 

```python
buffers = states_projected.centroid.buffer(distance=200000) #This is a 200 kilometer buffer
```

```python
fig, ax = plt.subplots(1, 1,figsize=(10,8))
states_projected.plot(ax=ax)
buffers.plot(ax=ax,color="green")
states_projected.centroid.plot(ax=ax,color="red")
```


<br></br>
<font size="3">Then let's calcuate the area and length of each state, and create two new columns in the states GeoDataFrame.</font>





```python
states_projected['state_area'] = states_projected.area
states_projected['state_length'] = states_projected.length
```


<br></br>
<font size="3">Let's find out which state has the largest area? Have a guess before running the command?</font>


```python
states_projected.sort_values('state_area', ascending=False)[['NAME','state_area']]
```


<br></br>
<font size="3">Let's find out which state has the largest length (circumference)? Have another guess?</font>

```python
states_projected.sort_values('state_length', ascending=False)[['NAME','state_length']]
```
